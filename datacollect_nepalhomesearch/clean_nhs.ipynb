```python
import csv
```

```python
#format to sq ft
csv_file = 'nepalhomessearch_dataset.csv'
updated_csv_file = 'nepalhomessearch_dataset1.csv'

def convert_to_square_feet(value):
    parts = value.split('-')
    if len(parts) == 4:
        try:
            ropani = int(parts[0])
            aana = int(parts[1])
            dam = int(parts[2])
            paisa = int(parts[3])

            square_feet = (ropani * 5476) + (aana * 342.25) + (dam * 21.39) + (paisa * 85.56)
            return str(square_feet)
        except ValueError:
            pass
    return value

updated_rows = []

with open(csv_file, 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    fieldnames = reader.fieldnames
    for row in reader:
        row['landsize'] = convert_to_square_feet(row['landsize'])
        updated_rows.append(row)

with open(updated_csv_file, 'w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(updated_rows)
```

```python
#anna to sqft
csv_file = 'nepalhomessearch_dataset1.csv'
updated_csv_file = 'nepalhomessearch_dataset2.csv'

def multiply_landsize(value):
    try:
        size = float(value)
        if size < 150:
            size *= 342.25
        return str(size)
    except ValueError:
        return value

updated_rows = []

with open(csv_file, 'r', encoding='utf-8', ) as file:
    reader = csv.DictReader(file)
    fieldnames = reader.fieldnames
    for row in reader:
        row['landsize'] = multiply_landsize(row['landsize'])
        updated_rows.append(row)

with open(updated_csv_file, 'w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(updated_rows)
```

```python
#parking 3+ 3-
csv_file = 'nepalhomessearch_dataset2.csv'
updated_csv_file = 'nepalhomessearch_dataset3.csv'
def update_parking_value(value):
    try:
        numeric_value = int(value)
        if numeric_value >= 3:
            return '3+'
        else:
            return '3-'
    except ValueError:
        return value

updated_rows = []

with open(csv_file, 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    fieldnames = reader.fieldnames
    for row in reader:
        row['parking'] = update_parking_value(row['parking'])
        updated_rows.append(row)

with open(updated_csv_file, 'w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(updated_rows)
```

```python
#date ad to bs
csv_file = 'nepalhomessearch_dataset3.csv'
updated_csv_file = 'nepalhomessearch_dataset4.csv'

def add_years(value):
    try:
        year = int(value)
        if year < 2024:
            year += 57
        return str(year)
    except ValueError:
        return value

updated_rows = []

with open(csv_file, 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    fieldnames = reader.fieldnames
    for row in reader:
        row['built year'] = add_years(row['built year'])
        updated_rows.append(row)

with open(updated_csv_file, 'w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(updated_rows)
```

```python
csv_file = 'nepalhomessearch_dataset4.csv'
updated_csv_file = 'nepalhomessearch_dataset6.csv'


def extract_district(data):
    if 'kathmandu' in data.lower():
        return 'kathmandu'
    elif 'lalitpur' in data.lower():
        return 'lalitpur'
    elif 'bhaktapur' in data.lower():
        return 'bhaktapur'
    else:
        return ''

updated_rows = []

with open(csv_file, 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    fieldnames = reader.fieldnames + ['district']
    for row in reader:
        location = row['location']
        title = row['title']
        district = extract_district(location) or extract_district(title)
        row['district'] = district
        updated_rows.append(row)

with open(updated_csv_file, 'w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(updated_rows)
```
