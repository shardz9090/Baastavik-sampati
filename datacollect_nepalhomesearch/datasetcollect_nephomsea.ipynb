```python
path = "C:\Program Files\Google\Chrome\Applicationchromedriver.exe"
service = Service(executable_path=path)
driver = webdriver.Chrome(service=service)
```

```python
with open('house_links_nephomsea.csv') as csv_file:
    csv_reader = csv.reader(csv_file)
    urls = [row[0] for row in csv_reader]
print(urls[200])
```

```python
data_list=[]
start_time = time.time()

for url in urls:
    try:
        driver.get(url)
        data_dict = {}
        try:
            title = driver.find_element(By.CSS_SELECTOR, value = 'body > div.rh_wrap.rh_wrap_stick_footer > section.rh_section.rh_wrap--padding.rh_wrap--topPadding > div > div.rh_page__head.rh_page__property > div.rh_page__property_title > h1')
            data_dict['title'] = title.text            
        except: pass
        
        try:
            location = driver.find_element(By.CSS_SELECTOR, value ='body > div.rh_wrap.rh_wrap_stick_footer > section.rh_section.rh_wrap--padding.rh_wrap--topPadding > div > div.rh_page__head.rh_page__property > div.rh_page__property_title > p')
            data_dict['location'] = location.text       
        except: pass
        
        try:
            price=driver.find_element(By.CSS_SELECTOR, value ='body > div.rh_wrap.rh_wrap_stick_footer > section.rh_section.rh_wrap--padding.rh_wrap--topPadding > div > div.rh_page__head.rh_page__property > div.rh_page__property_price > p.price')    
            data_dict['price'] = price.text
        except: pass
        
        try:
            area = driver.find_element(By.CSS_SELECTOR, value ='#property-content-section-overview > div.rh_property__row.rh_property__meta_wrap > div.rh_property__meta.prop_lot_size > div > span.figure')
            area2 = driver.find_element(By.CSS_SELECTOR, value ='#property-content-section-overview > div.rh_property__row.rh_property__meta_wrap > div.rh_property__meta.prop_lot_size > div > span.label')
            data_dict['land size'] = area.text+ " " + area2.text
            
        except: pass
        
        try:
            builty = driver.find_element(By.CSS_SELECTOR, value='#property-content-section-overview > div.rh_property__row.rh_property__meta_wrap > div.rh_property__meta.prop_year_built > div > span')
            data_dict['built year'] = builty.text
        except: pass
        
        try: 
            direction = driver.find_element(By.CSS_SELECTOR, value='#property-content-section-overview > div.rh_property__row.rh_property__meta_wrap > div.rh_property__meta.orientation- > div > span')
            data_dict['direction'] = direction.text
        except: pass
        
        try: 
            roaddis = driver.find_element(By.CSS_SELECTOR, value='#property-content-section-overview > div.rh_property__row.rh_property__meta_wrap > div.rh_property__meta.road > div > span')
            data_dict['Road'] = roaddis.text
        except: pass
        
        try:
            bedroom = driver.find_element(By.CSS_SELECTOR, value='#property-content-section-overview > div.rh_property__row.rh_property__meta_wrap > div.rh_property__meta.prop_bedrooms > div > span')
            data_dict['bedroom'] = bedroom.text    
        except: pass
        
        try:
            bathroom = driver.find_element(By.CSS_SELECTOR, value='#property-content-section-overview > div.rh_property__row.rh_property__meta_wrap > div.rh_property__meta.prop_bathrooms > div > span')
            data_dict['bathroom'] = bathroom.text        
        except: pass
        
        try:
            parking = driver.find_element(By.CSS_SELECTOR, value='#property-content-section-overview > div.rh_property__row.rh_property__meta_wrap > div.rh_property__meta.prop_garages > div > span')
            data_dict['parking size'] = parking.text
        except: pass
        
        data_dict['url']=url
        # Print the dictionary
        data_list.append(data_dict)
    except: pass

end_time = time.time()
execution_time = end_time - start_time
print("done")
print(f"Execution time: {execution_time}")
```
