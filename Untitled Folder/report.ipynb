```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('latlong.csv')
dataset.head()

#All prices are in Nepali Rupees and Land area in square feet
```

```python
report = dataset[['Price', 'Land']]
```

```python
average_price = report['Price'].mean()
average_land = report['Land'].mean()
print("Average Price:", average_price)
print("Average Land Area:", average_land)
```

```python
price_data = dataset['Price']
land_data = dataset['Land']
```

```python
#histogram of price

plt.figure(figsize=(10, 6))
plt.hist(price_data, bins=10, color='blue', edgecolor='black')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Price')
plt.grid(True)
plt.show()
```

```python
#histogram of land

plt.figure(figsize=(10, 6))
plt.hist(land_data, bins=10, color='red', edgecolor='black')
plt.xlabel('Land')
plt.ylabel('Frequency')
plt.title('Histogram of Land')
plt.grid(True)
plt.show()
```

```python
#scatter plot

plt.figure(figsize=(8, 6))
plt.scatter(land_data, price_data, color='purple', alpha=0.6)
plt.xlabel('Land')
plt.ylabel('Price')
plt.title('Scatter Plot: Price vs Land')
plt.grid(True)
plt.show()
```

```python
mean_price = dataset['Price'].mean()
median_price = dataset['Price'].median()
min_price = dataset['Price'].min()
max_price = dataset['Price'].max()

report = f"""
Data Analysis Report:

- Number of data points: {len(dataset)}
- Land area summary:
    - Minimum land area: {dataset['Land'].min()}
    - Maximum land area: {dataset['Land'].max()}
- House price summary:
    - Minimum house price: {min_price}
    - Maximum house price: {max_price}
    - Average house price: {mean_price}
    - Median house price: {median_price}
"""

print(report)
```
