```python
# land size 0-0-0-0 to sqft
def convert_to_square_feet(value):
    parts = value.split(',')
    square_feet = 0.0
    for part in parts:
        sub_parts = part.split('-')
        
        if len(sub_parts) == 4:
            try:
                aana = int(sub_parts[1])
                ropani = int(sub_parts[0])
                dam = int(sub_parts[3])
                paisa = int(sub_parts[2])

                square_feet += (aana * 342.25) + (ropani * 5476) + (dam * 21.39) + (paisa * 85.56)
                break
            except ValueError:
                pass
    return square_feet

csv_file = 'updated_houses8.csv'

with open(csv_file, 'r') as file:
    reader = csv.DictReader(file)
    rows = list(reader)

for row in rows:
    landsize = row['landsize']
    converted = convert_to_square_feet(landsize)  
   
    if converted ==0:
        print(landsize)
        row['landsize'] = landsize
    else:
        row['landsize'] = str(converted)

with open('new2w.csv', 'w', newline='') as file:
    fieldnames = reader.fieldnames
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(rows)
```

```python
# parking 3+ 3-
csv_file = 'updated_houses2.csv'
updated_csv_file = 'updated_houses3.csv'

def update_parking_value(value):
    try:
        numeric_value = int(value)
        if numeric_value >= 3:
            return '3+'
        else:
            return '3-'
    except ValueError:
        return value

updated_rows = []

with open(csv_file, 'r') as file:
    reader = csv.DictReader(file)
    fieldnames = reader.fieldnames
    for row in reader:
        row['parking'] = update_parking_value(row['parking'])
        updated_rows.append(row)

with open(updated_csv_file, 'w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(updated_rows)
```

```python
#landsize aana to sqft
csv_file = 'updated_houses6.csv'
updated_csv_file = 'updated_houses7.csv'

def convert_to_square_feet(value):
    try:
        parts = value.split()
        if len(parts) == 2 and parts[1].upper() == 'AANA':
            aana = float(parts[0])
            return str(aana * 342.25)
    except (ValueError, IndexError):
        pass
    return value

updated_rows = []

with open(csv_file, 'r') as file:
    reader = csv.DictReader(file)
    fieldnames = reader.fieldnames
    for row in reader:
        row['landsize'] = convert_to_square_feet(row['landsize'])
        updated_rows.append(row)

with open(updated_csv_file, 'w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(updated_rows)
```

```python
#convert landsize to 0-0-0-
csv_file = 'updated_houses7.csv'
updated_csv_file = 'updated_houses8.csv'

def convert_to_ropani_aana_daam_paisa(value):
    try:
        parts = value.split()
        if any(keyword.upper() in parts for keyword in ['AANA', 'PAISA', 'DAM', 'ROPANI']):
            aana = 0
            ropani = 0
            dam = 0
            paisa = 0

            for part in parts:
                if part.upper() == 'AANA':
                    aana = int(parts[parts.index(part) - 1])
                elif part.upper() == 'PAISA':
                    paisa = int(parts[parts.index(part) - 1])
                elif part.upper() == 'DAM':
                    dam = int(parts[parts.index(part) - 1])
                elif part.upper() == 'ROPANI':
                    ropani = int(parts[parts.index(part) - 1])

            return f'{ropani}-{aana}-{dam}-{paisa}'

    except (ValueError, IndexError):
        pass
    return value

updated_rows = []

with open(csv_file, 'r') as file:
    reader = csv.DictReader(file)
    fieldnames = reader.fieldnames
    for row in reader:
        row['landsize'] = convert_to_ropani_aana_daam_paisa(row['landsize'])
        updated_rows.append(row)

with open(updated_csv_file, 'w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(updated_rows)
```

```python
#remove rented houses
csv_file = 'updated_houses91.csv'
updated_csv_file = 'updated_houses92.csv'

def remove_rent_rows(rows):
    updated_rows = []
    for row in rows:
        if 'rent' not in row['title'].lower():
            updated_rows.append(row)
    return updated_rows

rows = []

with open(csv_file, 'r') as file:
    reader = csv.DictReader(file)
    fieldnames = reader.fieldnames
    rows = list(reader)

updated_rows = remove_rent_rows(rows)

with open(updated_csv_file, 'w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(updated_rows)

```
