```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from time import sleep
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import csv
import pandas as pd
from selenium.webdriver.chrome.options import Options
```

```python
path = "..\chromedriver.exe"
service = Service(executable_path=path)
driver = webdriver.Chrome(service=service)
driver.get('https://www.gharbazar.com/property/search/?_q=&_vt=0&_r=0&_pt=both&_si=0&_srt=latest')
driver.maximize_window()
```

```python
with open('linkgb.csv') as csv_file:
    csv_reader = csv.reader(csv_file)
    urls = [row[0] for row in csv_reader]
```

```python
data_list=[]
start_time = time.time()

for url in urls:
    try:
        driver.get(url)
        data_dict = {}
        try:
            address = driver.find_element(By.CSS_SELECTOR, '#property_details > div > div > div.row > div:nth-child(2) > div > ul > li:nth-child(1) > strong')
            data_dict['Address'] = address.text
        except: pass
        try:
            land = driver.find_element(By.CSS_SELECTOR, '#property_details > div > div > div.row > div:nth-child(1) > div > ul:nth-child(2) > li:nth-child(1) > strong')
            data_dict['Land'] = land.text
        except: pass
        try:
            road = driver.find_element(By.CSS_SELECTOR, '#property_details > div > div > div.row > div:nth-child(2) > div > ul > li:nth-child(2) > strong')
            data_dict['Road'] = road.text
        except: pass
        try:
            price = driver.find_element(By.CSS_SELECTOR, '#property-image > div > div.property-page__information > div.amount > h2')
            data_dict['Price'] = price.text
        except: pass
        try:
            bedroom = driver.find_element(By.CSS_SELECTOR, '#property_details > div > div > div.row > div:nth-child(1) > div > ul:nth-child(2) > li:nth-child(5) > strong')
            data_dict['Bedroom'] = bedroom.text
        except: pass
        try:
            bathroom = driver.find_element(By.CSS_SELECTOR, '#property_details > div > div > div.row > div:nth-child(1) > div > ul:nth-child(2) > li:nth-child(8) > strong') 
            data_dict['Bathroom'] = bathroom.text    
        except: pass
        try:
            direction = driver.find_element(By.CSS_SELECTOR, '#property_details > div > div > div.row > div:nth-child(1) > div > ul:nth-child(4) > li:nth-child(1) > strong')
            data_dict['Face'] = direction.text
        except: pass
        try:
            floor = driver.find_element(By.CSS_SELECTOR, '#property_details > div > div > div.row > div:nth-child(1) > div > ul:nth-child(2) > li:nth-child(9) > strong')
            data_dict['Floors'] = floor.text
        except: pass
        
        data_dict['url']=url
        data_list.append(data_dict)
    except: continue

end_time = time.time()
execution_time = end_time - start_time
print("done")
print(f"Execution time: {execution_time}")
```

```python
import pandas as pd

df = pd.DataFrame(data_list)
df.to_csv('housesgbnb.csv',index=False)
df_input = pd.read_csv('housesgbnb.csv')
df_input.tail()
```
