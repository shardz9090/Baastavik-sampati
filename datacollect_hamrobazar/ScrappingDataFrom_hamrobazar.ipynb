```python
import os, time, csv, re
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import pandas as pd
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# Create ChromeOptions object
options = Options()

# Disable notifications
options.add_argument("--disable-notifications")
path = '../chromedriver.exe'
```

```python
#loading Driver
service = Service(executable_path=path)
driver = webdriver.Chrome(service=service)
```

```python
# Load the CSV file and extract the URLs
with open('sub_dataframe4.csv') as file:
    urls = file.readlines()
    urls = [url.strip() for url in urls]
urls[1]
```

```python
data_list = []
start_time = time.time()

for url in urls:
    try:
        driver.get(url)
        data_dict = {}

        # Wait for the price element to be visible
        price = WebDriverWait(driver, 2).until(EC.visibility_of_element_located((By.XPATH, '//*[@id="hb__root"]/div[2]/main/div[2]/aside[1]/div[1]/section/div[2]/h2')))
        data_dict['price'] = price.text

        # Wait for the general container element to be visible
        general_container = WebDriverWait(driver, 2).until(EC.visibility_of_element_located((By.XPATH, '//*[@id="hb__root"]/div[2]/main/div[2]/section/div[3]/div/div[2]')))
        general_child = general_container.find_elements(By.XPATH, './div/div')

        for child in general_child:
            general_key = child.find_element(By.XPATH, './/div[1]').text
            general_value = child.find_element(By.XPATH, './/div[2]').text
            data_dict[general_key] = general_value
            print(general_key + ":" + general_value)

        # Wait for the specification container element to be visible
        specification_container = WebDriverWait(driver, 2).until(EC.visibility_of_element_located((By.XPATH, '//*[@id="hb__root"]/div[2]/main/div[2]/section/div[3]/div/div[3]')))
        specification_child = specification_container.find_elements(By.XPATH, './div/div')

        for child in specification_child:
            specification_key = child.find_element(By.XPATH, './/div[1]').text
            specification_value = child.find_element(By.XPATH, './/div[2]').text
            data_dict[specification_key] = specification_value
            print(specification_key + ':' + specification_value)

        data_dict['url'] = url
        print(data_dict)
        data_list.append(data_dict)
    except: pass

end_time = time.time()
print(data_list)
```

```python

# Create a DataFrame from the data_list
df = pd.DataFrame(data_list)

# Specify the desired CSV file path
csv_file_path = 'hamrobazar7.csv'

# Export the DataFrame to a CSV file
df.to_csv(csv_file_path, index=False)

print("CSV file created successfully.")
```

```python
driver.get(urls[0])
# Find the parent <ul> element containing the <li> elements using the CSS selector
ul_element = driver.find_element(By.CSS_SELECTOR, '#property_details > div > div > div.row > div:nth-child(1) > div > ul:nth-child(2)')

# Find all <li> elements within the <ul>
li_elements = ul_element.find_elements(By.TAG_NAME, "li")

# Create an empty dictionary to store the key-value pairs
data_dict = {}

# Iterate over each <li> element
for li_element in li_elements:
    # Extract the text from the <i> element
    key = li_element.find_element(By.TAG_NAME, "i").get_attribute("class")
    
    # Extract the text from the <strong> element
    value = li_element.find_element(By.TAG_NAME, "strong").text
    
    # Store the key-value pair in the dictionary
    data_dict[key] = value

# Print the dictionary
print(data_dict)
```
